```python
# Sección 1.6: Plotting with matplotlib.pyplot

# Gráfica en un panel
import matplotlib.pyplot as plt # importar matplotlib para poder hacer los gráficos
import numpy as np # aunque ya se importó atrás, de nuevo numpy
from numpy import sin, cos, tan, absolute

x = np.arange(-4, 4, 0.2) # Crear el eje x del gráfico. ¿Cómo se ajusta el eje x y sus puntos?

# Crear el gráfico
plt.plot(x, x*cos(x), 'y-', x, -x*cos(x), 'k-',x, absolute(x), ':r',x, -absolute(x), ':b', x, x*0, '-')
plt.xlabel('x'); plt.ylabel('f(x)')
plt.title('Gráfica de varias funciones')
plt.legend(('$xcos(x)$', '$-xcos(x)$', '$\mid x\mid$', '$-\mid x\mid$'),loc = 0) 
# como en matlab 0 es para la mejor ubicación

plt.grid(False)
plt.show()
```

```python
# Sección 1.6: Plotting with matplotlib.pyplot

# Gráfica con varios paneles: transformaciones de una función
import matplotlib.pyplot as plt
from numpy import arange,sin,cos
x = arange(-7,7,0.01)
plt.subplot(4,1,1)
plt.plot(x,sin(1/x),'-k')
plt.xlabel('x');plt.ylabel('sin(x)')
plt.grid(True)
plt.title('Transformaciones de sin(1/x)')
plt.subplot(4,1,2)
plt.plot(x,x*sin(1/x),'-.b')
plt.xlabel('x');plt.ylabel('x*sin(x)')
plt.grid(True)
plt.subplot(4,1,3)
plt.plot(x,x**2*sin(1/x),'--r')
plt.xlabel('x');plt.ylabel('$x^{2}sin(1/x)$')
plt.grid(True)
plt.subplot(4,1,4)
plt.plot(x,x**3*sin(1/x),':g')
plt.xlabel('x');plt.ylabel('$x^{3}sin(1/x)$')
plt.grid(True)
plt.show()
# Es necesario corregir el eje de las abscisas de x en cada subgráfico (menos el 4). 
```
