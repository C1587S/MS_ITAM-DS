# Introducción a ggplot2

```r
library(ggplot2)
options(repr.plot.width=4, repr.plot.height=4) #esta línea sólo se ejecuta para jupyterlab con R

# nota, la mayoría de estos ejemplos son tomados del libro "elegant graphics for data analysis" de Wickham(2016)
```

La finalidad de ggplot es generar gráficos en capas que mejoran la calidad y brindan mayor flexibilidad a los gráficos que R genera en su base.

```r
ggplot(mpg, aes(x = displ, y = hwy)) +
  geom_point()
```

```r
ggplot(mpg, aes(displ, hwy)) +
  geom_point()
```

```r
ggplot(mpg, aes(displ, cty, colour = class)) +
  geom_point()
```

```r
ggplot(mpg, aes(displ, hwy)) + geom_point(aes(colour = "blue"))
ggplot(mpg, aes(displ, hwy)) + geom_point(colour = "blue")
```

```r
sp<-ggplot(mtcars, aes(x=wt, y=mpg, color=cyl)) + geom_point()
sp
```

dentro de las capas anteriores se utilizó principalmente la de geom_point(), la cual permite presentar los puntos en forma de puntos

Es posible agreggar smoother a los geom_points

```r
ggplot(mpg, aes(displ, hwy)) +
  geom_point() +
  geom_smooth() 
```

loess es el método estandar para un número pequeño de n en el ds

En este ejemplo se utiliza como smoother el modelo lineal, con el método lm

```r
   ggplot(women, aes(height, weight)) +
     geom_point() +
     geom_smooth(method = "lm")
```

también es posible hacer boxplots y jittered plots para poder ver variación entre datos (generando ruido)


```r
ggplot(mpg, aes(drv, hwy)) +
  geom_point()
```

Para este ejemplo se presentan cpas que ayudan a evitar que los datos se trasnponga(jitter), diagramas de caja, y diagramas de violin(para ver la mayor concentración de los datos)

```r
ggplot(mpg, aes(drv, hwy)) + geom_jitter()
ggplot(mpg, aes(drv, hwy)) + geom_boxplot()
ggplot(mpg, aes(drv, hwy)) + geom_violin()
```

Es posible generar una capa de una gráfica, guardarla y después añadir otra o se imprime, por ejemplo, acá se crea la capa p primero y luego se añade un geom_point.



```r
p <- ggplot(mpg, aes(displ, hwy, colour = factor(cyl)))
```

```r
p+ geom_point()
```

Otro tipo de capa que se puede añadir es la de histogram 

```r
ggplot(mpg, aes(hwy)) + geom_histogram(bins=15)
```

ggplot también tiene una función para generar gráficos rápidos de visualización, quickplot, o qplot

```r
qplot(displ, hwy, data = mpg, colour = "blue")
qplot(displ, data = mpg)
# este gráfico se creó arriba pero con ggplot
```

También se pueden aprovechar las ventajas de que algunos data sets estén divididos por grupos, por ejemplo acá

```r
data(Oxboys, package = "nlme")
```

```r
ggplot(Oxboys, aes(Occasion, height)) +
  geom_boxplot()
```

Otra característica importante de ggplot es que puede asignar colores por clase en "aes". En particular, mapea los valores de una variable contínua. Por ejemplo, note el gráfico que ncorpora la opción fill y el que no.
También se puede asignar una variable de grupo con group

```r
ggplot(mpg, aes(class)) +
  geom_bar()
ggplot(mpg, aes(class, fill = drv)) +
  geom_bar()
ggplot(mpg, aes(class, fill = hwy, group = hwy)) +
  geom_bar()
```
