# Aproximación a una función por el teorema de Taylor¶
## Ejercicio 1

Graficar la función y los polinomios constante, lineal y cuadrático en una sola gráfica con matplotlib en el intervalo $\left[0,1\right]$. ¿Cuánto es la aproximación de los polinomios en $x$= 1.9? Calcula el error relativo de tus aproximaciones.





```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, cos, tan, absolute

x_ax = np.arange(1,2.01,0.01)

x_cent = 1.5

# La siguiente función evalúa vectores
def p_con(x, x0):
    pol_con = 2/3 + x*0;
    return pol_con
def p_lin(x, x0):
    pol_lin = 2/3 - (1/x0**2)*(x-x0)
    return pol_lin
def p_cua(x, x0):
    pol_cua = 2/3 - (1/x0**2)*(x-x0) + (2/x0**3)*(x-x0)**2
    return pol_cua

# Crear el entorno del gráfico
plt.plot(x_ax, 1/x_ax, 'k-') # Función original (1/x)
plt.plot(x_ax, p_con(x_ax, x_cent), 'g-', x_ax, p_lin(x_ax, x_cent), 'b:', x_ax, p_cua(x_ax, x_cent), 'r-.')
plt.plot(1.5, 1/1.5, 'Dc')
plt.plot(1.9, 1/1.9, 'Xc') # Este punto para comparar los polinomios y errores relativos de abajo
plt.xlabel('x'); plt.ylabel('f(x)')
plt.title('Función original y polinomios de Taylor',fontweight="bold")

plt.legend(('$f(x)=1/x$', 'Pol. Constante', 'Pol. Lineal', 'Pol. Cuadratico', 'x = 1.5, f(x)=2/3', 'x = 1.9, f(x) = 1/1.9'),loc = 0) 

plt.grid(True)
plt.show()

# Computo del error relativo
def rel_err(p, p_star):
    re = absolute(p - p_star)/absolute(p_star)
    return re

p_s = 1/1.9
x_re = 1.9

print('<'*10, 'Errores Relativos', '>'*5)
print('-'*10)
print(f"La función original f(x) = 1/x en x = 1.9 es: {1/1.9}")
print('-'*10)
re_con = rel_err(p_con(x_re, x0), p_s)
print(f"El P. constante evaluado en x = {x_re} es {p_con(x_re, x0)}")
print(f"El error relativo del P. constante es: {re_con}")
print('-'*10)
re_lin = rel_err(p_lin(x_re, x0), p_s)
print(f"El P. Lineal evaluado en x = {x_re} es {p_lin(x_re, x0)}")
print(f"El error relativo del P. Lineal es: {re_lin}")
print('-'*10)
re_cua = rel_err(p_cua(x_re, x0), p_s)
print(f"El P. Cuadrático evaluado en x = {x_re} es {p_cua(x_re, x0)}")
print(f"El error relativo del P. Cuadrático es: {re_cua}")
print('-'*10)
```
