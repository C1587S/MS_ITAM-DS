# Diferenciación numérica
## Ejercicio 2

Realizar una gráfica de $log\left(error\,relativo\right)$ vs $log\left(h\right)$ (h en el eje horizontal) para aproximar la segunda derivada de: $f(x)=e^{-x}$ en $x=1$ con $h \in \{10^{-16}, 10^{-14}, \dots , 10^{-1}\}.$ Valor a aproximar: $f^{(2)}(1) = e^{-1}$.

```python
import dif_hacia_delante as derivada
from numpy import absolute, exp, log
import numpy as np
np.set_printoptions(precision=16) # precisión de floats en numpy
```

```python
# Computo del error relativo
def err_re(p, p_star):
    re = absolute(p - p_star)/absolute(p_star)
    return re
```

```python
# Funciones a evaluar y constantes
f = lambda x: exp(-x)
x = 1
aux = np.arange(-16, 0, 1)
h = np.power(10.0, aux)
objetivo_df = -exp(-x)
objetivo_ddf = exp(-x)
```

```python
# Computo de las aproximaciones y los errores relativos
aprox_df = derivada.aprox_derivada_hacia_delante(f, x, h)
re_df = err_re(aprox_df, objetivo_df)
hdf_ast_pos = np.argmin(re_df)
hdf_ast = h[hdf_ast_pos]
aprox_ddf = derivada.aprox_2a_derivada_hacia_delante(f, x, h)
re_ddf = err_re(aprox_ddf, objetivo_ddf)
hddf_ast_pos = np.argmin(re_ddf)
hddf_ast = h[hddf_ast_pos]
```

```python
# Error relativo de Primeras derivadas en x=1
import matplotlib.pyplot as plt

axes = plt.gca()
#plt.legend(('Aprox. 1ra derivada(h)'), loc = 0)
plt.plot(log(h), log(re_df), 'b-', log(hdf_ast), log(hdf_ast), 'Xr')
plt.legend(('log(err)', '$h^{*}$'), loc = 0)
plt.xlabel('log(h)'); plt.ylabel('Log (error relativo)')
plt.title('Aproximación de la 1ra derivada evaluada en x=1',fontweight="bold")
plt.grid(True)

plt.show()
print(f"El h óptimo es {hdf_ast}, esto es en escala logarítmica: {log(hdf_ast)}")
```

```python
# Error relativo de Segundas derivadas en x=1
import matplotlib.pyplot as plt

axes = plt.gca()
#plt.legend(('Aprox. 1ra derivada(h)'), loc = 0)
plt.plot(log(h), log(re_ddf), 'b-', log(hddf_ast), log(hddf_ast), 'Xr')
plt.legend(('log(err)', '$h^{*}$'), loc = 0)
plt.xlabel('log(h)'); plt.ylabel('Log (error relativo)')
plt.title('Aproximación de la 2da derivada evaluada en x=1',fontweight="bold")
plt.grid(True)

plt.show()
print(f"El h óptimo es {hddf_ast}, esto es en escala logarítmica: {log(hddf_ast)}")
```
