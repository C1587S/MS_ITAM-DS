### Ejercicio 1: 
Realizar un ejercicio de reconstrucción de imágen usando 10, 20, 30, 40 y 50 número de vectores singulares izquierdos, derechos y valores singulares.

$\clubsuit$**Tomado de las notas de clase:**<br>
Si $rank(A) = r$ entonces $A = \displaystyle \sum_{i=0}^r \sigma_i u_i v_i^T$ con $u_i$ $i$-ésima columna de U y $v_i$ $i$-ésima columna de V.
$\clubsuit$<br><br>

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_ori = plt.imread('panther.png')
img = img_ori[:,:,0]
```

```python
plt.subplot(1,2,1)
plt.imshow(img_ori)
plt.title('Imagen original', fontweight="bold")

plt.subplot(1,2,2)
imgplot = plt.imshow(img, cmap='gray')
plt.title('Imagen es escala de grises', fontweight="bold")
plt.show()
```

```python
# Función para calcular SDV truncada
def svd_trunc(U, S, V_T, r):
    row_img, col_img= img.shape 
    A_trunc = np.empty([row_img, col_img]) 
    for i in range(1, r+1):
        img_svd_i = np.outer((U[:,i]*S[i]),V_T[i,:])
        A_trunc += img_svd_i
    return A_trunc
```

```python
r_list = [1, 10, 20, 30, 40, 50]
U,S,V_T = np.linalg.svd(img, full_matrices=False)

for i in r_list:
    img_svd_i = svd_trunc(U, S, V_T, i)
    
    plt.imshow(img_svd_i, cmap='gray')
    plt.title(f"SVD truncada a {i}", fontweight="bold")
    plt.show()
```
